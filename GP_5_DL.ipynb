## Оценивание и штрафы



Групповой проект оценивается исходя из устной защиты (презентации) командой выполненного ими проекта перед комиссией. Комиссия составляется из преподавателей курса, ассистентов курса, а также приглашенных экспертов из индустрии.

В рамках каждой защиты комиссия выставляет два типа оценок.

Первая оценка — это оценка для всего проекта в целом. Она ставится комиссией, исходя из глубины проработки командой предметной области, сложности применяемых в проекте инструментов, а также прочих аспектов, связанных с содержанием выполняемого в проекте ТЗ. При выставлении этой оценки комиссия ориентируется на критерии, более подробно представленные ниже в разделе "Критерии оценки проекта".

Вторая оценка — это индивидуальная оценка каждого участника команды. Именно эта оценка выставляется впоследствии в ведомость. Индивидуальная оценка участника команды определяется комиссией, исходя из того, насколько участник ориентируется в выполненном командой проекте, как он отвечает на адресованные персонально ему вопросы комиссии, насколько в принципе прослеживается его роль в проекте, а также исходя из других аспектов, связанных с участием конкретного человека в деятельности команды. Индивидуальная оценка не может превосходить оценку команды за проект в целом (первую оценку), однако, в случае когда участник спокойно и корректно отвечает на все адресованные ему вопросы, свободно ориентируется в материале и содержании выполненного проекта, а также когда его роль в работе команды не вызывает сомнений, — в таком случае, индивидуальная оценка участника становится равной оценке для всего проекта (т.е. в идеальном случае — вторая оценка равна первой оценке). Кроме того, в исключительных случаях, при демонстрации участником на защите превосходящих ожидания уровней понимания, участия и вовлечения в рамках проекта, комиссия может своим решением повысить индивидуальную оценку для этого участника на 0,5 или 1 балл относительно общей оценки за проект (первой оценки).

Во избежание субъективности и/или предвзятости оценки тех или иных членов комиссии в отношении тех или иных команд, а также каких-либо иных проявлений неравных условий — распределение команд между комиссиями осуществляется абсолютно рандомно с использованием средств библиотеки Numpy :)) — в качестве рэндом сида выступает дата, в которую производится распределение, например, `np.random.seed(25052026)`.

Обратите, пожалуйста, внимание на то, что бОльшая часть комиссий осуществляет приём проектов в очной форме, однако могут присутствовать и 1-2 комиссии, которые будут работать дистанционно с использованием средств видеосвязи. В случае попадания команды путем случайного распределения в комиссию, работающую в очной форме, *не допускается* какое-либо изменение формата проведения для команды на онлайн-режим, равно как и любые другие перемещения команд между комиссиями (смены комиссий), для сохранения честности и непредвзятости случайного распределения.

Явка студента на защиту проекта является обязательной; вклад студента без его присутствия не оценивается и автоматически является равным нулю. Для дистанционных комиссий дополнительным требованием является обязательное наличие включенной камеры в течение всего времени нахождения в конференции

## Описание задания

В современной бизнес-индустрии использование глубинного обучения приобретает критическое значение. Компании активно внедряют нейронные сети для решения самых разных задач: от анализа структурированных табличных данных до работы с изображениями и текстами. Это позволяет бизнесу принимать более обоснованные решения и оперативно реагировать на изменения рынка.

В данном проекте вы на реальных данных попрактикуетесь в построении и сравнении нейросетевых архитектур для решения двух бизнес-задач разной природы. Первая задача будет связана с табличными данными и решается с помощью полносвязных нейронных сетей. Вторая - с неструктурированными данными: изображениями или текстами, - и решается с помощью более сложных архитектур глубинного обучения. Помимо этого, вы получите опыт организации работы над ML-проектом так, как это принято в индустрии: с версионированием кода, воспроизводимыми экспериментами и сохранением всех артефактов.

Принципиально важно, что обе бизнес-задачи должны быть **содержательно связаны между собой** и **относиться к одной и той же компании или платформе** — таким образом, чтобы вместе они образовывали единое и логически целостное решение для выбранного бизнес-контекста. Проект, в котором две задачи решаются для разных компаний или никак не связаны между собой, оценивается в **0 баллов** для всех участников команды.

### Этапы проекта

Ниже вашему вниманию предлагаются примерные этапы реализации проекта:

- Выбор компании и/или платформы;
- Формулировка двух связанных бизнес-задач, которые могла бы хотеть решить компания/платформа на основе данных: одна на табличных данных, другая на неструктурированных (изображения или текст);
- Обоснование сформулированных бизнес-задач и описание того, как предлагаемое решение вписывается в бизнес-процессы компании;
- Получение необходимых данных путём поиска открытых источников/датасетов, парсинга/работы с API;
- Формулировка задач в терминах глубинного обучения для решения поставленных бизнес-задач;
- Выбор и обоснование архитектур нейронных сетей для каждой из задач;
- Первичная обработка данных;
- Построение, обучение и сравнение нейросетевых архитектур с обязательным логированием экспериментов;
- Получение и агрегация результатов, финальные выводы.

### Требования к проекту

В качестве результата выполнения группового проекта ваша команда должна получить презентацию и защитить её перед комиссией, а также предоставить ссылку на репозиторий GitHub с кодом.

Оформление презентации остаётся полностью на ваше усмотрение, но помните, что результат должен быть релевантен для демонстрации бизнес-заказчику — комиссию, принимающую вашу работу, правильнее всего воспринимать именно в таком качестве. Например, вставлять в презентацию строчки кода или злоупотреблять скринами блокнота не рекомендуется.

С точки зрения концепции выполнения проекта, вам необходимо принять на себя роль дата-инженеров: продумать идеологию получения и извлечения данных; дата-аналитиков: соотнести бизнес-задачу с формулировками на языке данных; и главное, безусловно, — дата-сайентистов: реализовать поставленные задачи с помощью методов глубинного обучения и извлечь полезные инсайты из проделанной работы. А в конце, конечно, презентовать всю эту красоту в понятном виде потенциальному заказчику.

Разумеется, чем сложнее и чем нетривиальнее будет ваша задача, а также чем глубже и чем осмысленнее будут выводы, извлечённые в рамках этой задачи, — тем выше окажется качество вашего проекта и тем выше он будет оценён!

## Критерии оценки проекта

Оценка за групповой проект в целом — то есть первая из двух оценок — выставляется по 10-балльной шкале, исходя из 5 критериев. Также комиссия на своё усмотрение может добавить (но не снизить) некоторое количество бонусных баллов за рамками данных критериев.

Каждый из критериев будет перечислен и рассмотрен ниже.

### Критерий 1. Выбор задач и бизнес-обоснование (до 2 баллов)

Данный критерий оценивает, насколько обоснованными и осмысленными и **содержательно связанными** получились две решаемые в проекте бизнес-задачи; насколько полно был осуществлён переход от бизнес-задач к задачам глубинного обучения (насколько в реальности постановка задач соответствует исходной постановке задачи с точки зрения интересов компании); а также насколько корректными, объёмными и прикладными для бизнеса получились выявленные в ходе исследования инсайты.

По данному критерию можно получить **максимальный балл** только при условии, что в презентации представлена схема, иллюстрирующая, как предлагаемое решение вписывается в бизнес-процессы компании: откуда поступают данные, каким образом они обрабатываются, как модель генерирует предсказания и каким образом эти предсказания используются для принятия бизнес-решений. Строгая нотация не обязательна — важна логика и понятность схемы для стороннего наблюдателя.

### Критерий 2. Инструменты извлечения и обработки данных (до 1,5 баллов)

Данный критерий оценивает техническую реализацию процесса сбора данных ручным и автоматизированным образом, использованные средства предобработки, подготовки данных и EDA. В рамках этого критерия будет рассматриваться спектр различных технических инструментов, внедрённых в проект; корректность, эффективность и правильность их применения; а также сложность и новизна алгоритмов, подходов и библиотек, применяемых к данным.

### Критерий 3. Глубинное обучение на табличных данных (до 2 баллов)

Данный критерий оценивает применение полносвязных нейронных сетей для решения бизнес-задачи на табличных данных. В рамках этого критерия будет рассматриваться корректность постановки задачи, осмысленность выбора архитектуры, качество предобработки данных, а также глубина сравнительного анализа различных конфигураций сети.

*__Важно:__ в случае **оценки 0** за этот критерий оценка за весь проект становится равна оценке за этот критерий.*

*По данному критерию можно получить **отличное от нуля количество баллов** только при условии корректного применения полносвязной нейронной сети для решения поставленной задачи на табличных данных.*

*По данному критерию можно получить **1 балл и выше** только при условии осуществления сравнительного анализа различных конфигураций и сложности архитектуры: должно быть рассмотрено не менее трёх вариантов, значимо отличающихся друг от друга — например, по глубине сети, размеру скрытых слоёв, функциям активации, наличию регуляризации. Использование каждой конфигурации, как и подбор гиперпараметров, должны быть подробно описаны и обоснованы; в рамках сравнения должны быть получены и сформулированы явные, значимые выводы о том, как изменение архитектуры влияет на качество решения задачи.*

*По данному критерию можно получить **максимальный балл в 2 балла** только при условии корректного и обоснованного применения архитектурных решений или подходов, не рассматривавшихся в рамках соответствующих лекций и семинаров. Кроме того, **максимальный балл в 2 балла** по данному критерию недостижим без логирования экспериментов в соответствии с требованиями **Критерия 5**. При этом ожидается, что каждая из сравниваемых конфигураций залогирована отдельным запуском с сохранением параметров архитектуры и обучения, метрик качества, а также артефакта обученной модели, позволяющего воспроизвести её предсказания без повторного обучения.*

### Критерий 4. Глубинное обучение на неструктурированных данных  (до 3 баллов)

Данный критерий оценивает применение методов глубинного обучения для решения бизнес-задачи на неструктурированных данных: изображениях или текстах. В рамках этого критерия будет рассматриваться спектр различных технических инструментов, внедрённых в проект; корректность, эффективность и правильность их применения; а также сложность и новизна алгоритмов, подходов и библиотек, применяемых к данным. Кроме того, будет оценен сравнительный анализ архитектур нейронных сетей, использованных в решении.

*__Важно:__ в случае **оценки 0** за этот критерий оценка за весь проект становится равна оценке за этот критерий.*

*По данному критерию можно получить **отличное от нуля количество баллов** только при условии использования в проекте сверточных нейронных сетей, рекуррентных нейронных сетей или же более сложных архитектур глубоких нейронных сетей.*

*По данному критерию можно получить **2 балла** только при условии осуществления сравнительного анализа эффективности различных архитектур нейронных сетей внутри вашего проекта на одних и тех же данных. Самих архитектур должно быть не менее трёх; они должны значимо и осмысленно друг от друга отличаться; их использование, как и настройка гиперпараметров, должны быть подробно описаны и обоснованы. Кроме того, в рамках сравнительного анализа должны быть получены и сформулированы явные, значимые выводы о сравнительной эффективности той или иной архитектуры.*

*По данному критерию можно получить **максимальный балл в 3 балла** только при условии корректного и обоснованного применения инструментов/подходов/библиотек для построения моделей, не рассматривавшихся в рамках соответствующих лекций и семинаров. Кроме того, **максимальный балл в 3 балла** по данному критерию недостижим без логирования экспериментов в соответствии с требованиями **Критерия 5**. При этом ожидается, что каждая из сравниваемых архитектур залогирована отдельным запуском с сохранением параметров архитектуры и обучения, метрик качества в разрезе эпох, а также артефакта обученной модели, позволяющего воспроизвести её предсказания без повторного обучения.*

### Критерий 5. Инструменты промышленной разработки (до 1,5 баллов)

Данный критерий оценивает, насколько грамотно и профессионально команда организовала процесс совместной разработки и обеспечила воспроизводимость своих экспериментов. В рамках этого критерия рассматриваются две составляющие: оформление репозитория GitHub и логирование экспериментов.

*По данному критерию можно получить **0,5 балла** при выполнении всех следующих требований к репозиторию GitHub:*

- наличие файла `README.md` с описанием проекта, используемых данных и структуры репозитория — после прочтения этого файла у стороннего человека не должно остаться вопросов о том, что делает проект и как в нём разобраться;

#Бизнес составляющая проекта

**Компания:** Медицинская клиника

**Бизнес задача 1:**
Атоматизация скрининга пациента для проверки наличия пневмонии на основе симптомов

**Бизнес задача 2:**
Автоматизация проверки рентгеновских снимков грудной клетки пациента для выявления пневмонии


**Обоснование бизнес задач:**
В период сезонных заболеваний в поликлиники обращается множество пациентов с признаками респираторных заболеваний. Для быстрого выявления тяжелого заболевания - пневмонии, необходима автоматизация этого процесса.

**Решение:**
Внедрение двухэтапной системы принятия решения:
1. Модель оценивает вероятность наличия пневмонии по клиническим симптомам для необходимости прохождения рентгена легких
2. Модель оценивает рентгеновский снимок грудной клетки пациента для помощи определения наличия пневмонии

**Связь задач:**
Первая задача используется для первичного скрининга пациента, вторая после прохождения рентгена. В совокупности они образуют полный цикл выявления пневмонии у пациента

**Польза:**
- Ускорение процесса диагностики
- Сокращение нагрузки на медицинский персонал
- Быстрое выявление пациентов с пневмонией
- Более точное назначение рентгена, за счет чего идет уменьшение вредного влияния этого на больных


**Схема работы в процессах компании:**
Пациент обращается в поликлинику -> ему отправляется специальная форма, в которой предлагается заполнить данные о его симптомах -> оценка вероятности наличия пневмонии -> назначение рентгена -> анализ снимка -> при подтверждении этого заболевания - приоритетная запись на следующий прием к врачу -> итоговое решение врача и назначение лечения

```python
#строим первую модель
class first_net(nn.Module): # наследуемся от класса nn.Module
    def __init__(self):
        super(first_net,self).__init__()
        hidden_1 = 256
        hidden_2 = 128

        self.net = torch.nn.Sequential(
                      torch.nn.Linear(23, hidden_1),
                      torch.nn.ReLU(),
                      torch.nn.Linear(hidden_1, hidden_2),
                      torch.nn.Linear(hidden_2, 2)
                    )

    def forward(self,x):
        x = self.net(x)
        return x
```

```python
# функция обучения модели
def train(model, device, train_loader, optimizer, criterion, epoch, WANDB):
    model.train() # обязательно переводим в режим обучения
    test_loss = 0
    correct = 0

    n_ex = len(train_loader)

    for batch_idx, (data, target) in tqdm(enumerate(train_loader), total=n_ex):
        data, target = data.to(device), target.to(device) # переводим картинки и таргеты на GPU
        # обнуляем градиенты!
        optimizer.zero_grad()
        # прямой проход
        output = model(data)
        pred = output.argmax(dim=1, keepdim=True)
        correct += pred.eq(target.view_as(pred)).sum().item() # определяем корректность предсказания
        train_loss = criterion(output, target) # считаем значение функции потерь
        # обратный проход
        train_loss.backward()
        # делаем градиентный шаг оптимизатором
        optimizer.step()
        # считаем метрики и лосс
    tqdm.write('\nTrain set: Average loss: {:.4f}, Accuracy: {:.0f}%'.format(
        train_loss, 100. * correct / len(train_loader.dataset)))

    if WANDB:
        wandb.log({'train_loss': train_loss,
                   'train_accuracy': correct / len(train_loader.dataset)})
```

```python
# функция тестирования
def test(model, device, test_loader, criterion, WANDB):
    model.eval() # переводем модель в режим инференса
    test_loss = 0
    correct = 0
    # показываем, что обученич нет и градиенты не обновляются
    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss = criterion(output, target) # считаем значение функции потерь
            pred = output.argmax(dim=1, keepdim=True)
            correct += pred.eq(target.view_as(pred)).sum().item()
            # считаем метрики
    tqdm.write('Test set: Average loss: {:.4f}, Accuracy: {:.0f}%'.format(
        test_loss, 100. * correct / len(test_loader.dataset)))

    if WANDB:
        wandb.log({'test_loss': test_loss,
                   'test_accuracy': correct / len(test_loader.dataset)})
```

```python
# основная функция для экспериментов
def main(model):
    # настраиваем окружение для wandb
    if CFG.wandb:
        os.environ["WANDB_API_KEY"] = CFG.api
        wandb.init(project=CFG.project, entity=CFG.entity, reinit=True, config=class2dict(CFG))

    use_cuda = torch.cuda.is_available() # проверяем доступность GPU

    seed_everything(CFG.seed) # фиксируем сиды

    device = torch.device("cuda" if use_cuda else "cpu") # выделили устройство

    if CFG.wandb:
        wandb.watch(model, log='all') # логируем все (метрики, лоссы, градиенты)

    for epoch in range(1, CFG.num_epochs + 1): # цикл на эпохи
        print('\nEpoch:', epoch)
        train(model, device, train_loader, optimizer, criterion, epoch, CFG.wandb)
        test(model, device, test_loader, criterion, CFG.wandb)
    print('Training is ended!')
```

```python
#строим вторую модель
#документация, которую использовали https://docs.pytorch.org/docs/2.12/nn.html
class second_net(nn.Module): # наследуемся от класса nn.Module
    def __init__(self):
        super(second_net,self).__init__()
        hidden_1 = 256
        hidden_2 = 128

        self.net = nn.Sequential(
            nn.Linear(23, hidden_1),
            nn.BatchNorm1d(hidden_1),
            nn.ReLU(),
            nn.Dropout(0.1),

            nn.Linear(hidden_1, hidden_2),
            nn.BatchNorm1d(hidden_2),
            nn.ReLU(),
            nn.Dropout(0.1),

            nn.Linear(hidden_2, 2)
        )
    def forward(self,x):
        x = self.net(x)
        return x
```

### Обучим третью модель

У нас есть несколько признаков, которые хорошо коррелируют с целевой переменной, поэтому если научить модель выбирать наилучшие признаки при помощи механизма внимания, это позволит улучшить ее качество, так как она будет лучше обучаться на более важном признаке.

```python
# основная функция для экспериментов
def main(model):
    # настраиваем окружение для wandb
    if CFG.wandb:
        os.environ["WANDB_API_KEY"] = CFG.api
        wandb.init(project=CFG.project, entity=CFG.entity, reinit=True, config=class2dict(CFG))

    use_cuda = torch.cuda.is_available() # проверяем доступность GPU

    seed_everything(CFG.seed) # фиксируем сиды

    device = torch.device("cuda" if use_cuda else "cpu") # выделили устройство

    if CFG.wandb:
        wandb.watch(model, log='all') # логируем все (метрики, лоссы, градиенты)

    for epoch in range(1, 40): # цикл на эпохи
        print('\nEpoch:', epoch)
        train(model, device, train_loader, optimizer, criterion, epoch, CFG.wandb)
        test(model, device, test_loader, criterion, CFG.wandb)
    print('Training is ended!')
```

```python
# строим третью модель с механизмом внимания
#документация про механизм внимания https://habr.com/ru/articles/1020624/
#в качестве функции softmax из статьи взяли функцию тангенса
class third_net(nn.Module):
    def __init__(self):
        super(third_net, self).__init__()
        self.attention = nn.Sequential(
            nn.Linear(23, 23),
            nn.Tanh(),
            nn.Linear(23, 23)
        )
        self.fc = nn.Sequential(
            nn.Linear(23, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 2)
        )

    def forward(self, x):
        attn_weights = torch.softmax(self.attention(x), dim=1)
        weighted_x = x * attn_weights
        return self.fc(weighted_x)
```

```python
model = third_net()
print(model)
```

```python
# функция потерь (cross-entropy)
criterion = nn.CrossEntropyLoss()

# SGD (классический) со скоростью обучения (learning rate) = 0.01
optimizer = torch.optim.SGD(model.parameters(),lr = 0.01)
```

```python
main(model)
```

Увеличили количество эпох, так как заметили постепенное уменьшение ошибки с каждой эпохой без переобучения. По итогу получили accuracy 97% с уровнем ошибки 0.1432
